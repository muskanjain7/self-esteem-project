#### Column categorization

```python
column_categorization = {
    'SRH' : {
        'nutrition' : ['N1','N2','N3','N4','N5'], 
        'parental_connection' : ['P1','P2','P3','P4'], 
        'decision_making' : ['D1','D2','D3','D4'], #'D5_imp','D6_imp'based on marital status
        'sp_harassment' : ['S1','S2','S3','S4','S5'],
        'mhm' : ['M1','M2','M3','M4','M5','M6',],
        'pi_sa' : ['PS1','PS2','PS3','PS4'],
        'body_image' : ['B1','B2','B3','B4','B5']
    }, 
    'demographics' : ['Age', 'Edu Status', 'Marital Status']
}

# with open(os.path.join(os.path.dirname(os.path.realpath("__file__")), "data", "column_categorization.pickle"), 'wb') as handle:
#     pickle.dump(column_categorization, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
# with open(os.path.join(os.path.dirname(os.path.realpath("__file__")), "data", "column_categorization.pickle"), 'rb') as handle:
#     column_categorization = pickle.load(handle)
```

```python
def add_columns(df):
    df["nutrition_score"] = df[column_categorization["SRH"]['nutrition']].sum(axis=1)  
    df["parental_connection_score"] = df[column_categorization["SRH"]['parental_connection']].sum(axis=1)   
    df["decision_making_score"] = df[column_categorization["SRH"]['decision_making']].sum(axis=1) + np.where(df["Marital Status"]=="Married",df["D5"],df["D6"])
    df["sp_harassment_score"] = df[column_categorization["SRH"]['sp_harassment']].sum(axis=1)  
    df["mhm_score"] = df[column_categorization["SRH"]['mhm']].sum(axis=1)
    df["pi_sa_score"] = df[column_categorization["SRH"]['pi_sa']].sum(axis=1)
    df["body_image_score"] = df[column_categorization["SRH"]['body_image']].sum(axis=1)
    df["srh_score"] = (
        df["nutrition_score"] + 
        df["decision_making_score"] +
        df["parental_connection_score"] + 
        df["sp_harassment_score"] +
        df["mhm_score"] +
        df["pi_sa_score"] +
        df["body_image_score"] 
    )
    
    return df
```

```python
df1 = add_columns(df1)
df2 = add_columns(df2)
df3 = add_columns(df3)
df4 = add_columns(df4)
dfa = add_columns(dfa)
dfb = add_columns(dfb)
dfc = add_columns(dfc)
```
